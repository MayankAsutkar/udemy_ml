### Indentation

#### Indentation in py is used to define the structure and hierarchy of the code.Unlike many other programming languages that use braces {} to delimit blocks of code,Py uses indentation to determine the grouping of statements.This means that all the statements within a block must be indented at the sam level.  

```python
## Indentation

age = 32
if  age > 30:
    print(age)
    
print(age)
```

```python
## line continuation

total = 1+2+3+4+5+6+7+\
    4+5+6
    
print(total)
```
