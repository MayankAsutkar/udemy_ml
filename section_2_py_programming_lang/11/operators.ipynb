```python
## Arithmetic operations

a = 10
b = 5

add_res = a + b
sub_res = a - b
pro_res = a * b
div_res = a / b
floor_div_res = a // b
mod_res = a%b
expo_res = a**b
print(add_res)
print(sub_res)
print(pro_res)
print(div_res)
print(floor_div_res)
print(mod_res)
print(expo_res)
```
