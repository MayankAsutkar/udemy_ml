```python
## sets

## sets are a built-in data type in python used to store collections of unique items.unordered elements.duplicate elemenst not allowed

## iterate a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set((1,2,3,4,5,6))
print(type(my_empty_set))
print(my_empty_set)
```

```python
## Basics sets operation
##adding and removing elements
my_set.add(7)
print(my_set)
my_set.add(7)
print(my_set)
```

```python
## remove the elements from a set

my_set.remove(3)
print(my_set)
```

```python
# my_set.remove(10) will give error
```

```python
my_set.discard(11)
print(my_set)
```

```python
## pop method 
removed_element = my_set.pop()
print(removed_element)
my_set
```

```python
my_set.clear()
print(my_set)
```

```python
## set memebership test
my_set = {1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## mathematics operation

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## union
union_set = set1.union(set2)
union_set

## intersection
intersection_set = set1.intersection(set2)
intersection_set

set1.intersection_update(set2)
print(set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## difference'
print(set1.difference(set2))
```

```python
set2.difference(set1)
```

```python
## symmetric Difference
set1.symmetric_difference(set2)
```

```python
## set Methods

set1 = {1,2,3}
set2 = {3,4,5}

## is subset

print(set1.issubset(set2))

print(set1.issuperset((set2)))
set1.issuperset(())
```

```python
## counting unique words in text

text = "In this tutorial we are discussung about sets"
words = text.split()

## convert list of words to set to get unique words

unique_words = set(words)
print(unique_words)
```
