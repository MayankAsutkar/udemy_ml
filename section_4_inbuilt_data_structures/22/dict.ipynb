```python
## dictionaries
"""
    unordered
    store data in key-value pairs
    keys are immutable
    value can be any type
"""
```

```python
## creating dictionaries

empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student = {"name":"krish","age":32,"grade":24}
print(student)
print(type(student))
```

```python
## sungle key is always used
student = {"name":"krish","age":32,"name":24}
print(student)
```

```python
## accessing dictionary elements
student = {"name":"krish","age":32,"grade":'A'}
print(student)
```

```python
## accessing dictionary elements

print(student['grade'])
print(student['age'])

### accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))

## default values
print(student.get('last_name',"Not avaialable"))
```

```python
## modifying Dictionary elements
## dictionary are mutable ,add,update and delete
print(student)
```

```python
student['age'] = 33 ## updated values of the key
print(student)
student["address"] = "India"    ## added a new key and value
print(student)
```

```python
del student['grade'] ## delete key and value pair

print(student)
```

```python
## dictionary methods 

keys = student.keys()
print(keys)
values = student.values()
print(values)

items = student.items()
print(items)
```

```python
## shallow copy
student_copy = student ## any changes made in shallow copy will also be made in original
print(student)
print(student_copy)
```

```python
student['name'] = "krsih1"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() ## this is a shallow copy
student_copy1["name"] = "krish3"
print(student)
print(student_copy1)
```

```python
### iterating over dictionaries
## using loops

## iterating over keys

for keys in student.keys():
    print(keys)
```

```python
## iterating over values

for value in student.values():
    print(value)
```

```python
## itearting over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## nested dictionaries

students={
    "student1":{"name":"krish","age":32},
    "student2":{"name":"peter","age":35}
}
print(students)
```

```python
## accessing nested dictionaries elements

print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
## iteratung over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## dictionarry comprehension

squares = { x:x**2 for x in range(5)}
print(squares)
```

```python
## cnditional ditionary comprehension

evens = {x:x**2 for x in range(10) if x%2 == 0}
print(evens)
```

```python
## practical examples

## use q dictionary to count frequency of elements in list

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency={}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
```

```python
## merge 2 dictinaries

dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
