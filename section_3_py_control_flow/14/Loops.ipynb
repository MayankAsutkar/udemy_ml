```python
## Loop control statements

## break
## The break statement : 

for i in range(10):
    if i==5:
        break
    print(i)
    
    
```

```python
# control statements:

## The continue statement skips the current iteration and continues with the next

for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
## pass 
## The pass statement is a null operation; it does nothing

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## nested loops

# A loop inside a loop.

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## examples :
# calculate the sum of 1 st  n natural nunmbers:
n = int(input("Enter the val of n"))
sum = 0
while n!=0:
    sum+=n
    n-=1
print(sum)
```

```python
## prime numbers:

for num in range(1,10):
    if num>1:
        for i in range(2,num):
            if num%i == 0:
                break
        else:
            print(num)
```
