```python
## map : gives a function to all items in an input list and return a map object. 

numbers = [1,2,3,4,5,6,7]
list(map(lambda x:x*x,numbers))
```

```python
## map takes multiple iterables

number1=[1,2,3]
number2=[4,5,6]

added_numbers = list(map(lambda x,y : x+y,number1,number2))
print(added_numbers)
```

```python
## mapp to convert a list of strings to integers

str_num = ['1','2','3','4','5']
int_num = list(map(int,str_num))
print(int_num)
```

```python
def get_name(person):
    return person['name']

people = [
    {'name':'krish','age':32},
    {'name':'Jack','age':33}
]

list(map(get_name,people))
```
