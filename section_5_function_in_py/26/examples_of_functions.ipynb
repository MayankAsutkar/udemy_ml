```python
def convert_temp(temp,unit):
    """converts temp between celsius and fahrenheit"""
    if unit == 'C':
        return temp * 9/5 + 32 ## converts celsius to fahrenheit
    elif unit == 'F':
        return (temp - 32)*5/9 ## fahrenheit to celsius
    else:
        return None
    
print(convert_temp(25,'C'))
print(convert_temp(77,'F'))
        
```

```python
## password strength checker

def pswd_check(pswd):
    """Checks if password is  strong"""
    if len(pswd) < 8:
        return False
    if not any(char.isdigit() for char in pswd):
        return False
    if not any(char.islower() for char in pswd):
        return False
    if not any(char.isupper() for char in pswd):
        return False
    if not any(char in '!@#$%^&*()_+' for char in pswd):
        return False
    return True
```

```python
## caliing the functions

print(pswd_check("Weakpswd"))
print(pswd_check("Weakpswd@12"))
```

```python
## calculate total cost of items in a shopping cart

def calc_cost(cart):
    total_cost = 0
    for item in cart:
        total_cost += item['Price']*item['quantity']
    return total_cost

## example of cart data:

cart = [
    {'name':'Apple','Price':0.5,'quantity':4},
    {'name':'Banana','Price':0.3,'quantity':6},
    {'name':'Orange','Price':0.7,'quantity':3}
]


## calling the function
print(calc_cost(cart))
```

```python
## check if a string is pallindrome

def is_pallindrome(s):
    s = s.lower().replace(" ","")
    return s == s[::-1]

print("A man a plan a canal panama")
print(is_pallindrome("A man a plan a canal panama"))
print(is_pallindrome("A man a plan a canal panam"))
```

```python
## fqctorial calculations

def factorial(n):
    if n==0:
        return 1
    else:
        return n* factorial(n-1)
    
print(factorial(6))
```

```python
## to read a file and count the frewuncy of each word

def count_word(file_path):
    word_count = {}
    with open(file_path,'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!?;:"\'')
                word_count[word] = word_count.get(word,0)+1
    return word_count

filepath = 'sample.txt'
word_frequency = count_word(filepath)    
print(word_frequency)
```
