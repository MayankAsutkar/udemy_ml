```python
def add(a,b):
    return a+b

add(2,3)
```

```python
## syntax: lambda arguments : expression

add = lambda a,b : a+b

type(add)
```

```python
print(add(5,6))
```

```python
def even(num):
    if num%2 == 0:
        return True
    
even(24)
```

```python
even = lambda num:num%2 == 0
even(12)
```

```python
def add(x,y,z):
    return x+y+z

add(1,2,3)
```

```python
add = lambda x,y,z: x+y+z
add(1,2,4)
```

```python
## map() : applies a function to all items in a list

numbers = [1,2,3,4,5,6]
def square(number):
    return number**2
print(square(4))
```

```python
list(map(lambda x:x**2,numbers))
```
